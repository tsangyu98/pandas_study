```python
# 导包
import numpy as np
import pandas as pd
```

```python
# 作业1：如图创建一个Dataframe（5*2，值为0-100的随机值），并分别计算key1和key2的均值、中位数、累积和
df = pd.DataFrame({'key1': np.random.rand(5) * 100,
                  'key2': np.random.rand(5) * 100})
print('创建DataFrame为:\n', df)
print('-' * 88)
print('df["key1"]的均值为:\n', df['key1'].mean())
print('-' * 88)
print('df["key1"]的中位数为:\n', df['key1'].median())
print('-' * 88)
print('df["key2"]的均值为:\n', df['key2'].mean())
print('-' * 88)
print('df["key2"]的中位数:\n', df['key2'].median())
print('-' * 88)
df['key1_cumsum'] = df['key1'].cumsum()
df['key2_cumsum'] = df['key2'].cumsum()
print('df的累计和:\n', df)
```

```python
# 作业2：写出一个输入元素直接生成数组的代码块，然后创建一个函数，该函数功能用于判断一个Series是否是唯一值数组，返回“是”和“不是”
s = input("请随机输入一组元素，用逗号(英文符号)隔开:\n")
s_list = s.split(',')
sr = pd.Series(s_list)
if len(s_list) == len(sr.unique()):
    print('该数组是唯一数组')
else:
    print('该数组不是唯一数组')
```

```python
# 作业1：如图创建一个Dataframe，并分别通过字符串常用方法得到3个Series或得到新的Dataframe：
# ① name字段首字母全部大写
# ② gender字段去除所有空格
# ③ score字段按照-拆分，分别是math，english，art三个学分
df = pd.DataFrame({'gender':['M', 'M', 'F', 'M', 'F'],
                  'name':['Jack', 'Tom', 'Marry', 'Zack', 'Heheda'],
                  'score': ['90-91-92', '89-88-87', '90-90-90', '78-78-78', '60-60-60']})
print('创建DataFrame为:\n', df)
print('-' * 88)
df['name'] = df['name'].str.capitalize()
df['gender'] = df['gender'].str.strip()
df['math'] = df['score'].str.split('-').str[0]
df['english'] = df['score'].str.split('-').str[0]
df['art'] = df['score'].str.split('-').str[0]
del df['score']
print('修改后为:\n', df)
print('-' * 88)
```

```python
# 作业1：按要求创建Dataframe df1、df2，并合并成df3
df1 = pd.DataFrame({'key':list('abc'),
                   'values1': np.random.rand(3)})
print('创建df1为:\n', df1)
print('-' * 88)
df2 = pd.DataFrame({'key':list('bcd'),
                   'values2': np.random.rand(3)})
print('创建df2为:\n', df2)
print('-' * 88)
df3 = pd.merge(df1, df2, how='outer', on='key')
print('合并df3为:\n', df3)
```

```python
# 作业2：按要求创建Dataframe df1、df2，并合并成df3
df1 = pd.DataFrame({'lkey':list('abc'),
                   'values1': np.random.rand(3)})
print('创建df1为:\n', df1)
print('-' * 88)
df2 = pd.DataFrame({'rkey':list('bcd'),
                   'values2': np.random.rand(3)})
print('创建df2为:\n', df2)
print('-' * 88)
df3 = pd.merge(df1, df2, how='left', left_on='lkey', right_on='rkey')
print('合并df3为:\n', df3)
```

```python
# 作业3：按要求创建Dataframe df1、df2，并合并成df3
df1 = pd.DataFrame({'key':list('abc'),
                   'values1': np.random.rand(3)})
print('创建df1为:\n', df1)
print('-' * 88)
df2 = pd.DataFrame({'values2': np.random.rand(3),
                   'values3': [5, 6, 7]}, index=list('bcd'))

print('创建df2为:\n', df2)
print('-' * 88)
df3 = pd.merge(df1, df2, how='inner', left_on='key', right_index=True)
print('合并df3为:\n', df3)
```

```python
# 作业2：按要求创建Dataframe df1、df2，并用df2的值修补df1，生成df3
df1 = pd.DataFrame({'values1': [np.random.rand(), np.nan, np.nan, np.random.rand()],
             'values2': np.random.rand(4)}, index=list('abcd'))
print('创建df1为:\n', df1)
print('-' * 88)
df2 = pd.DataFrame({'values1': np.arange(0, 7, 2), 
                   'values2': np.arange(1, 8, 2)}, index=list('abcd'))
print('创建df2为:\n', df2)
print('-' * 88)
print('df1修补后:\n', df1.combine_first(df2))
```

```python
# 作业1：按要求创建Dataframe df，并通过分组得到以下结果
# ① 以A分组，求出C,D的分组平均值
# ② 以A,B分组，求出D,E的分组求和
# ③ 以A分组，得到所有分组，以字典显示
# ④ 按照数值类型分组，求和
# ⑤ 将C,D作为一组分出来，并计算求和
# ⑥ 以B分组，求出每组的均值，求和，最大值，最小值
df = pd.DataFrame({'A' : ['one', 'two', 'three', 'one','two', 'three', 'one', 'two'],
                   'B' : ['h', 'h', 'h', 'h', 'f', 'f', 'f', 'f'],
                   'C' : np.arange(10,26,2),
                   'D' : np.random.randn(8),
                   'E':np.random.rand(8)})
print('创建df为:\n', df)
print('-' * 88)
print('以A分组，求出C,D的分组平均值:\n', df.groupby('A').mean())
print('-' * 88)
print('以A,B分组，求出D,E的分组求和:\n', df.groupby(['A', 'B']).sum())
print('-' * 88)
print('以A分组，得到所有分组，以字典显示:\n', df.groupby('A').groups)
print('-' * 88)
print('按照数值类型分组，求和:\n', df.groupby(df.dtypes,axis=1).apply(lambda df:df.sum(axis=1)))
print('将C,D作为一组分出来，并计算求和:\n', df[['C', 'D']])
print('将C,D作为一组分出来，并计算求和:\n', df[['C', 'D']].sum(axis=1))
tdf = df[['C', 'D']].sum(axis=1)
tdf = pd.DataFrame({'r': tdf})
print('将C,D作为一组分出来，并计算求和:\n', tdf)
print('以B分组，求出每组的均值，求和，最大值，最小值:\n', df.groupby('B').agg([np.mean, np.sum, np.max, np.min]))
print('-' * 88)


```

```python
# 作业1：按要求创建Dataframe df，通过key分组求和，并将求和结果并在原df中
df = pd.DataFrame({'data1': np.random.rand(8),
             'data2': np.random.rand(8),
             'key': ['a', 'a', 'b', 'b', 'a', 'b', 'a', 'b']})
print('创建df为:\n', df)
print('-' * 88)
df2 = df.groupby('key').transform(np.sum)
print(df2)
print('-' * 88)
print(df.join(df2, rsuffix='sum_'))
```

```python
# 作业1：按要求创建Dataframe df，并通过数据透视表或交叉表得到以下结果
# ① 以A列聚合，求出C,D的平均值
# ② 以A,B列聚合，求出D,E的均值、求和
# ③ 以B聚合，计算A列的元素频率
df = pd.DataFrame({'A' : ['one', 'two', 'three', 'one','two', 'three', 'one', 'two'],
                   'B' : ['h', 'h', 'h', 'h', 'f', 'f', 'f', 'f'],
                   'C' : np.arange(10,26,2),
                   'D' : np.random.randn(8),
                   'E':np.random.rand(8)})
print('创建df为:\n', df)
print('*' * 88)
df2 = pd.pivot_table(df, index='A', values=['C', 'D'], aggfunc=np.mean)
print("以A列聚合，求出C,D的平均值:\n", df2)
print('*' * 88)
df3 = pd.pivot_table(df, index=['A', 'B'], values=['D', 'E'], aggfunc=[np.mean, np.sum])
print('以A,B列聚合，求出D,E的均值、求和:\n', df3)
print('*' * 88)
df4 = pd.crosstab(df['B'], df['A'])
print('以B聚合，计算A列的元素频率:\n', df4)
```
