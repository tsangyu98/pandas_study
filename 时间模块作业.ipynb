```python
# 作业1：请通过迭代创建一个时间列表（如图，时间区间为任意一个月，这里可以用for语句），然后转化成DatetimeIndex，并print月中是多少号
datelst = []
for i in range(1, 32):
    datelst.append('2017-12-%02d' % i)
print('创建时间列表为:', datelst)
print('---------------------------------------------')
import pandas as pd
t = pd.DatetimeIndex(datelst)
print('转换为DatetimeIndex为:', t)
print('---------------------------------------------')
n = (len(t)+1) / 2 - 1
print('月中日期为:', t[int(n)])
```

```python
# 作业2：请如图创建一个包含时间日期的txt文件，通过open语句读取后转化成DatetimeIndex
with open('date.txt', 'r') as f:
    s = f.readline()
# print(s)
data = s.split(',')
print(data)
print('-----------------------------------')
t = pd.to_datetime(data)
print('转化为DatetimeIndex:\n', t)
print('-------------------------------------')
```

```python
# 作业1：请输出以下时间序列
import numpy as np

t = pd.date_range('2017/1/1', '2017/1/5')
data = pd.Series(np.random.rand(len(t)), index=t)
print('时间序列1:\n', data)
print('--------------------------------------')
t = pd.date_range('2017', '2018', freq='3M')
data = pd.Series(np.random.rand(len(t)), index=t)
print('时间序列2:\n', data)
print('--------------------------------------')
t = pd.date_range('2017/12/1', '2017/12/1 00:30:00', freq='10MIN')
data = pd.DataFrame(np.random.rand(4, 4), index=t, columns=['value1', 'value2', 'value3', 'value4'])
print('时间序列3:\n', data)
```

```python
# 作业2：按要求创建时间序列ts1，并转换成ts2
t1 = pd.date_range('2017/5/1 12:00:00', '2017/5/1 12:40:00', freq='10T')
data = pd.Series(np.random.rand(len(t1)), index=t1)
print('创建时间序列ts1:\n', data)
print('--------------------------------------')
ts2 = data.asfreq('5T', 'ffill')
print(ts2)
```

```python
# 作业1：请输出以下时间序列，使用pd.period_range()
t1 = pd.period_range('2017/1', '2017/5', freq='M')
data = pd.Series(np.random.rand(len(t1)), index=t1)
print('时间序列1:\n', data)
t2 = pd.period_range('2017/1/1', periods=5, freq='2H')
data = pd.Series(np.random.rand(len(t2)), index=t2)
print()
print('时间序列2:\n', data)
```

```python
# 作业1：如图创建时间序列(10*3，值为0-100的随机数)，通过索引得到以下值
data = pd.DataFrame(np.random.rand(10, 3) * 100, index=pd.date_range('2017/12/1', periods=10, freq='12H'), columns=['value1', 'value2', 'value3'])
print(data)
print('----------------------------------------------------------')
print("前4行所有值为:\n", data[0:4])
print('----------------------------------------------------------')
print("2017-12-4 12:00:00的数据为:\n", data.loc["2017-12-4 12:00:00"])
print('----------------------------------------------------------')
print("2017-12-4 - 2017-12-5的数据:\n", data["2017-12-04": "2017-12-05"])
```

```python
# 作业1：按要求创建时间序列ts1，通过降采样和升采样，转换成ts2，ts3
ts1 = pd.Series(np.random.rand(10), index=pd.date_range('2017-1-1', periods=10))
print("创建时间序列ts1:\n", ts1)
print('----------------------------------------------------------')
ts2 = ts1.resample("3D")
print("转换成ts2:\n", ts2.mean())
print('----------------------------------------------------------')
ts3 = ts1.resample("12H")
print("转换成ts3:\n", ts3.ffill())
```
