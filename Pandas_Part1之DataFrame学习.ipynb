```python
# 导包
import numpy as np
import pandas as pd
```

```python
'''
Pandas数据结构DataFrame:基本概念及创建
"二维数组"DataFrame:是一个表格型的数据结构，包含一组有序的列， 其列的值类型可以是数值、字符串、布尔值等
DataFrame中的数据以一个或多个二维块存放，不是列表、字典或一维数组结构
'''
```

```python
# DataFrame数据结构
# DataFrame是一个表格型的数据结构，"带有标签的二维数组"
# DataFrame带有index(行标签)和columns(列标签)
data = {'name':['Jack', 'Tom', 'Marry'],
       'age':[18, 19, 20],
       'gender':['m', 'm', 'w']}
frame = pd.DataFrame(data)
print(frame)
print(type(frame))
print(frame.index,'该数据的类型为:', type(frame.index))
print(frame.columns, '该数据的类型为:', type(frame.columns))
print(frame.values, '该数据的类型为:', type(frame.values))
```

```python
# DataFrame创建方法一:由数组/list组成的字典
# 创建方法:pandas.DataFrame()
data1 = {'a':[1, 2, 3],
        'b':[3, 4, 5],
        'c':[6, 7, 8]}
df1 = pd.DataFrame(data1)
print(df1)
print('-------------------')
data2 = {'one':np.random.rand(3),
        'two':np.random.rand(3)}  # 如果这里填4会报错，列数必须都一样
df2 = pd.DataFrame(data2)
print(df2)
print('-------------------')
# 由数组/list组成的字典 创建Dataframe,columns为字典的key，index默认为数字标签

df1 = pd.DataFrame(data1, columns=['b','c', 'a', 'd'])
print(df1)
print('-------------------')
df1 = pd.DataFrame(data1, columns=['b','c'])
print(df1)
print('-------------------')
# columns参数:可以重新指定列的顺序，格式为list,如果现有数据中没有该列(比如'd'),则产生NaN值
# 如果columns重新指定时，列的数量可以少于原数据

df2 = pd.DataFrame(data2, index=['f1', 'f2', 'f3'])  # 3行数据，索引必须是3个，否则会报错
print(df2)
# index参数:重新定义index，格式为list，长度必须保持一致
```

```python
# DataFrame创建方法二:由Series组成的字典
data1 = {'one':pd.Series(np.random.rand(2)),
        'two':pd.Series(np.random.rand(3))}  
df1 = pd.DataFrame(data1)  # 没有设置index的Series
print(df1)
print('-----------------')
data2 = {'one':pd.Series(np.random.rand(2), index=['a', 'b']),
      'two':pd.Series(np.random.rand(3), index=['a', 'b', 'c'])}
df2 = pd.DataFrame(data2)
print(df2)
# 由Series组成的字典，创建DataFrame,columns为字典key,index为Series的标签(如果Series没有指定标签，则是默认数字标签)
# Series可以长度不一样，生成的DataFrame会出现NaN值
```

```python
# DataFrame创建方法三:通过二维数组直接创建
ar = np.arange(9).reshape(3, 3)
print(ar)
df1 = pd.DataFrame(ar)
print(df1)
df2 = pd.DataFrame(ar, index=['a', 'b', 'c'], columns=['one', 'two', 'three'])
print(df2)
# 通过二维数组直接创建DataFrame,得到一样形状的结果数据，如果不指定index和columns，两者均返回数字格式
# index和columns指定长度与原数组保持一致
```

```python
# DataFrame创建方法四：由字典组成的列表
data = [{'one':1, 'two':2}, {'one':5, 'two':10, 'three':20}]
df1 = pd.DataFrame(data)
print(df1)
df2 = pd.DataFrame(data, index=['a', 'b'])
print(df2)
df3 = pd.DataFrame(data, columns=['one', 'two'])
print(df3)
# 由字典组成的列表创建Dataframe,columns为字典中的key，index不做指定则为默认数子标签
# columns和index参数分别重新指定相应列及行标签
```

```python
# DataFrame创建方法五:由字典组成的字典
data = {'Jack':{'math':90, 'english':89, 'art':78},
       'Marry':{'math':89, 'english':66, 'art':79},
       'Tom':{'math':100, 'english':98, 'art':99}}
df1 = pd.DataFrame(data)
print(df1)
# 由字典组成的字典创建dataframe，columns为字典的columns，index为子字典的key
df2 = pd.DataFrame(data, columns=['Jack', 'Tom', 'Bob'])
print(df2)
df3 = pd.DataFrame(data, index=['a', 'b', 'c'])
print(df3)
# columns参数可以增加和减少现有列，如出现新的列，值为NaN
# index和之前不同，并不能改变原有index，如果指向新的标签，值为NaN
```

```python
'''
Pandas数据结构DataFrame:索引
DataFrame既有行索引也有列索引，可以被看作由Series组成的字典
选择行/选择列/切片/布尔判断
'''
```

```python
# 选择行与列
df = pd.DataFrame(np.random.rand(12).reshape(3, 4)*100, index=['one', 'two', 'three'], columns=['a', 'b', 'c', 'd'])
print(df)
print('-----------------------')
data1 = df['a']
print(data1, type(data1))
print('-----------------------')
data2 = df[['a', 'c']]
print(data2, type(data2))
print('-----------------------')
# 按照列名选择列，只选择一列的话是Series,选择多列的话是DataFrame

data3 = df.loc['one']
print(data3, type(data3))
print('-----------------------')
data4 = df.loc[['one', 'two']]
print(data4, type(data4))
# 按照index选择行，只选择一行是Series,选择多行的话是DataFrame
```

```python
# df[]-选择列
# 一般用于选择列，也可以选择行
df = pd.DataFrame(np.random.rand(12).reshape(3, 4)*100, index=['one', 'two', 'three'], columns=['a', 'b', 'c', 'd'])
print(df)
print('------------------')
data1 = df['a']
print(data1)
print('------------------')
data2 = df[['b','c']]
print(data2)
# df[]默认选择列,[]中写列名(所以一般数据columns都会单独制定，不会用默认的数字列名，以免和index冲突)
# 单选列为Series，print结果为Series
# 多选列为DataFrame，print结果为DataFrame
print('------------------')
data3 = df[:1]
print(data3, type(data3))
# df[]中为数字时，默认选择行，且只能进行切片的选择，不能单独选择
# 输出结果为DataFrame,即便只选择一行
# df[]不能通过索引标签名来选择行

# 核心笔记:df[col]一般用于选择列，[]中写列名
```

```python
# df.loc[] —— 按index选择行
df1 = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                  index=['one', 'two', 'three', 'four'],
                  columns=['a', 'b', 'c', 'd'])
print(df1)
print("---------------------------------")
df2 = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                  columns=['a','b', 'c', 'd'])
print(df2)
print("---------------------------------")
print(df1.loc['one'])
print('---------------------------------')
print(df2.loc[1])
print('---------------------------------')
# 单个标签索引，返回Series
data3 = df1.loc[['one', 'two', 'four']]
print(data3)
print('---------------------------------')
data4 = df2.loc[[2, 3]]
print(data4)
# 多个标签索引，如果标签不存在，则返回NaN，顺序可变
print('---------------------------------')
data5 = df1.loc['one':'three']
print(data5)
print('---------------------------------')
data6 = df2.loc[1:3]
print(data6)
# 切片索引:可以做切片对象
# 末端包含

# 核心笔记:df.loc[label]主要针对index选择行，同时支持指定index，及默认数字index
```

```python
# df.iloc[]——按照整数位置(从轴的0到length-1)选择行
# 类似list的索引，其顺序就是dataframe的整数位置，从0开始计
df = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                 index=['one', 'two', 'three', 'four'],
                 columns=['a','b','c','d'])
print(df)
print('---------------------------------')
print(df.iloc[0])
print('---------------------------------')
print(df.iloc[-1])
print('---------------------------------')
# 单位索引
# 和loc索引不同，不能索引超出数据行数的整数位置
print(df.iloc[[0,2]])
print(df.iloc[[3, 2, 1]])
# 多位置索引
# 顺序可变
print('---------------------------------')
print(df.iloc[1:3])
print(df.iloc[::2])
# 切片索引
# 末端不包含
```

```python
# 布尔型索引
# 和Series原理相同
df = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                 index=['one','two','three','four'],
                 columns=['a', 'b', 'c', 'd'])
print(df)
print('---------------------------------')
b1 = df < 20
print(b1, type(b1))
print('---------------------------------')
print(df[b1])
# 索引会对数据每个值进行判断，True返回原数据，False返回NaN
print('---------------------------------')
b2 = df['a'] < 50
print(b2, type(b2))
print(df[b2])
# 单列做判断，索引结果保留所有数据，单列判断为True的，包括其他列
print('---------------------------------')
b3 = df[['a', 'b']] > 50
print(b3, type(b3))
print(df[b3])
# 多列做判断
# 索引结果保留所有的数据:True返回原数据，False返回NaN
print('---------------------------------')
b4 = df.loc[['one', 'three']] < 50
print(b4, type(b4))
print(df[b4])
# 多行判断
# 索引结果保留所有数据:True保留原数据，False显示为NaN
```

```python
# 多重索引:比如同时索引行和列
# 先选择列在选择行——相当于对于一个数据先筛选字段，在选择数据量
df = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                 index=['one', 'two', 'three', 'four'],
                 columns=['a', 'b', 'c', 'd'])
print(df)
print('---------------------------------')
# 选择a列的one, three行
print(df['a'].loc[['one', 'three']])
print('---------------------------------')
# 选择b,c,d列的one,three行
print(df[['b', 'c', 'd']].loc[['one', 'three']])
print('---------------------------------')
# 选择满足判断索引的前两行数据
print(df[df['a'] > 50].iloc[:2])
```

```python
'''
Pandas数据结构:DataFrame:基本技巧
数据查看、转置/添加、修改、删除/对齐/排序
'''
```

```python
# 数据查看(和Series一样)、转置
df = pd.DataFrame(np.random.rand(16).reshape(8,2)*100,
                 columns=['a', 'b'])
print(df)
print('---------------------------------')
print(df.head(2))
print('---------------------------------')
print(df.tail())
print('---------------------------------')
# 默认查看5条信息
print(df.T)
# .T转置
```

```python
# 添加与修改
df = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                 columns=['a', 'b', 'c', 'd'])
print(df)
print('---------------------------------')
df['e'] = 10
print(df)
print('---------------------------------')
df.loc[4] = 20
print(df)
# 新增行/列
print('---------------------------------')

df['e'] = 20
print(df)
print('---------------------------------')
df[['a', 'c']] = 200
print(df)
# 索引后直接修改值
```

```python
# 删除
df = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                 columns=['a', 'b', 'c', 'd'])
print(df)
print('---------------------------------')
del df['a']
print(df)
# del语句用于删除列
print('---------------------------------')

print(df.drop(0))
print(df)
print('---------------------------------')
print(df.drop([1,2]))
print('---------------------------------')
# drop()删除行，inplace=False(默认),删除后生成新的数据，不改变原数据
print(df.drop(['d'], axis=1))
print(df)
# drop()删除列，需要加上axis=1,inplace=False:删除后生成的新的数据，不改变原数据
```

```python
# 对齐
df1 = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])
df2 = pd.DataFrame(np.random.randn(7, 3), columns=['A', 'B', 'C'])
print(df1 + df2)
# DataFrame对象之间的数据自动按照列和索引对齐
```

```python
# 排序1——按置排序 .sort_values
# 同样适用于Series
df1 = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                  columns=['a','b','c','d'])
print(df1)
print('---------------------------------')
print(df1.sort_values(['a'], ascending=True))  # 升序
print('---------------------------------')
print(df1.sort_values(['a'], ascending=False))  # 降序
print('---------------------------------')
# 单列排序

df2 = pd.DataFrame({'a':[1, 1, 1, 1, 2, 2, 2, 2],
                   'b':list(range(8)),
                   'c':list(range(8, 0, -1))})
print(df2)
print('---------------------------------')
print(df2.sort_values(['a', 'c']))
# 多列排序，按列排序
```

```python
# 排序2——索引排序 .sort_index
df1 = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                  index=[5, 4, 3, 2],
                  columns=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame(np.random.rand(16).reshape(4, 4)*100,
                  index=list('hsxg'),
                  columns=list('abcd'))
print(df1)
print('---------------------------------')
print(df2)
print('---------------------------------')
print(df1.sort_index())
print('---------------------------------')
print(df2.sort_index())
```
